# Regressors: orthogonalise RT regressors

```python
import numpy as np
from random import shuffle as sfl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import gamma
import scipy
import itertools
from scipy.linalg import sqrtm, inv
from scipy import stats


%matplotlib inline
```

### Define hrf functions 

```python
def spm_Gpdf(s,h,l):
    s = np.array(s)
    res = (h-1)*np.log(s) + h*np.log(l) - l*s - np.log(gamma(h))
    return np.exp(res)

def canonical(resolution):
    # translated from spm_hrf
    p=[6,16,1,1,6,0,32]
    dt = resolution/16.
    s = np.array(xrange(int(p[6]/dt+1)))
    #HRF sampled at 0.1 s
    hrf = spm_Gpdf(s,p[0]/p[2],dt/p[2]) - spm_Gpdf(s,p[1]/p[3],dt/p[3])/p[4]
    hrf = hrf[[int(x) for x in np.array(xrange(int(p[6]/resolution+1)))*16.]]
    hrf = hrf/np.sum(hrf)
    # duration of the HRF
    durhrf = 32.0
    # length of the HRF parameters in resolution scale
    laghrf = int(np.ceil(durhrf/resolution))
    hrf = hrf[:laghrf]

    return {"basishrf":hrf,"laghrf":laghrf}
```

### What does an orthogonalised RT regressor look like?

Orthogonalisation as in Jeanette's slides for the bootcamp:

$X_2 = \beta_1X_1 + \epsilon$


$X_2^{ortho} = X2 - \hat\beta_1X_1$


```python
X1 = np.zeros(300)
X1[range(50,100)+range(150,200)+range(250,300)]=1
reg_X1 = np.convolve(X1,canonical(1)['basishrf'][:len(X1)])
X2 = np.zeros(300)
X2[range(50,80)+range(150,220)+range(250,300)]=1
reg_X2 = np.convolve(X2,canonical(1)['basishrf'][:len(X2)])

fit = stats.linregress(reg_X1,reg_X2)
print("slope: %f"%fit.slope)
reg_X2_ortho = reg_X2-fit.slope*reg_X1

plt.figure(figsize=(12,8))
plt.plot(reg_X1,label="regressor 1, fixed duration")
plt.plot(reg_X2,label="regressor 2, RT duration")
plt.plot(reg_X2_ortho,label="regressor 2 orthogonalised with regressor 1")
plt.legend()
```

It would make sense if you could use the duration regressor as a correction for the difference in the modeled duration and the RT.

However, 2 questions:

1. Problem: There is still some leftover activation modeled during the fixed duration period of the trials in the orthogonalised regressor, depending on the parameter.  Moreover, the height of this activation depends on the subject's difference from the fixed duration. (and taking the fixed duration == mean RT does not solve this, see example)

2. Question: Is the goal of this orthogonalisation to 'control' for the difference in duration of each trial?  In that case, why not just use regressor 2, with duration == RT?


### Solution?

If we'd want something like below, why not immediately leave out regressor 1 and use regressor 2?

```python
# Is this the goal?
plt.figure(figsize=(12,8))
plt.plot(reg_X1,label="regressor 1, fixed duration")
plt.plot(reg_X2,label="regressor 2, RT duration")
plt.plot(reg_X1-reg_X2,label="regressor 2 orthogonalised with regressor 1")
plt.legend()
```
