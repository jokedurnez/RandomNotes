```python
# apparently it doesn't always success, so we need a function that retries until it works

def request_until_succeed(url):
    req = urllib2.Request(url)
    success = False
    while success is False:
        try: 
            response = urllib2.urlopen(req)
            if response.getcode() == 200:
                success = True
        except Exception, e:
            print(e)
            time.sleep(5)
            
            print("Error for URL %s: %s" % (url, datetime.datetime.now()))

    return response.read()
```

```python
# function to get get page information

def get_pagefeed(page_id,maxnum):
    access_token = get_access_token()
    
    # construct the URL string
    base = "https://graph.facebook.com/v2.4"
    node = "/" + page_id + "/feed" # changed
    parameters = "/?limit="+str(maxnum)+"&access_token=%s" % access_token
    url = base + node + parameters

    return url
```

```python
# Set up connection with facebook API

graph = facebook.GraphAPI(access_token=get_access_token(), version='2.2')
```

```python
# read in all the likes for the nytimes articles

page_id = 'nytimes'
totallikes = []
finished = False
url = get_pagefeed(page_id,20)
for x in range(1):
    pagefeed = json.loads(request_until_succeed(url))
    articles = pagefeed['data']
    url = pagefeed['paging']['next']

    for article in articles:
        #print('\n article likes: ',end='')
        id = article['id']
        likes = graph.get_object(id=id,fields='likes')['likes']
        finished = False
        articlelikes = []

        k = 0
        while (finished == False and k < 5):
            #print("=", end='')
            who = [x['id'] for x in likes['data']]
            articlelikes = articlelikes+who
            if 'next' in likes['paging'].keys():
                nextlikes = likes['paging']['next']
                likes = json.loads(request_until_succeed(nextlikes))       
            else:
                finished = True
            k = k+1
        totallikes.append(articlelikes)

# while statement
# here should go to next page as well !
```
